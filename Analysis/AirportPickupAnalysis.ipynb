```python
import pandas as pd
pd.set_option('display.mpl_style', 'default')
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
```

```python
airportdrop = pd.read_json("airportdropoffs.json", orient="index")
```

```python
airportdrop
```

```python
airportdrop[2010.0]=(airportdrop[2010.0]-airportdrop[2010.0].mean())/airportdrop[2010.0].std()
airportdrop[2011.0]=(airportdrop[2011.0]-airportdrop[2011.0].mean())/airportdrop[2011.0].std()
airportdrop[2012.0]=(airportdrop[2012.0]-airportdrop[2012.0].mean())/airportdrop[2012.0].std()
airportdrop[2013.0]=(airportdrop[2013.0]-airportdrop[2013.0].mean())/airportdrop[2013.0].std()
airportdrop[2014.0]=(airportdrop[2014.0]-airportdrop[2014.0].mean())/airportdrop[2014.0].std()
airportdrop[2015.0]=(airportdrop[2015.0]-airportdrop[2015.0].mean())/airportdrop[2015.0].std()
```

```python
airportdrop.dropna(inplace=True)
```

```python
airport95 = airportdrop[airportdrop<airportdrop.quantile(0.95)]
airport95.describe()
```

```python
airportdeltas = airport95.pct_change(axis = 1)
airportdeltas = airportdeltas.drop(2010.0, axis = 1)
airportdeltas.dropna()
airportdeltas.describe(percentiles=[0.05, 0.1, 0.5, 0.85, 0.9, 0.95])
```

```python
largechange = airportdeltas[airportdeltas>=airportdeltas.quantile(0.95)]
largechange = largechange.dropna(how='all')
largechange.describe()
```

```python
largechange.to_csv('highchangeairportwith20142015.csv')
```
